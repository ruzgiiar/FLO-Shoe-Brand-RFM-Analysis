### 3. Task (3. Görev): Calculation of RF Score (RF Skorunun Hesaplanması)

---

- **Step 1:** *Convert Recency, Frequency and Monetary metrics into scores between 1-5 with the help of qcut.*

  - (**Adım 1:** *Recency, Frequency ve Monetary metriklerini qcut yardımı ile 1-5 arasında skorlara çeviriniz.*)
 
---

- **Step 2:** *Save these scores as recency_score, frequency_score and monetary_score*

  - (**Adım 2:** *Bu skorları recency_score, frequency_score ve monetary_score olarak kaydediniz*)
 
---

- **Step 3:** *Express recency_score and frequency_score as a single variable and save as RF_SCORE*

  - (**Adım 3:** *recency_score ve frequency_score’u tek bir değişken olarak ifade ediniz ve RF_SCORE olarak kaydediniz*)
 
---

```python
rfm['recency_score'] = pd.qcut(rfm['recency'], 5, labels = [5, 4, 3, 2, 1])
```

```python
rfm['frequency_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm['monetary_socre'] = pd.qcut(rfm['monetary'], 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
rfm[[col for col in rfm.columns if rfm[col].dtypes in ['int64', 'float64']]].corr()
```

- **Due to the high correlation between Frequency and Monetary values, we prefer to use Frequency on the vertical axis and Recency on the horizontal axis in a two-dimensional segmentation for customer classification. This choice is made based on the strong relationship between shopping frequency and customer loyalty.**

    - *(Frequency ve Monetary değerleri arasında yüksek korelasyon olduğu için, müşteri segmentasyonu için 2 boyutlu bir sınıflandırmada dikey ekseni Frequency, yatay ekseni ise Recency değerlerini kullanmayı tercih ediyoruz. Bu seçim, alışveriş sıklığı ile müşteri sadakati arasındaki güçlü ilişki nedeniyle yapılmıştır.)*

```python
rfm['RF_SCORE'] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))
```

### 4. Task (4. Görev) : Defining RF Score as a Segment (RF Skorunun Segment Olarak Tanımlanması)

- **Step 1:** *Make segment definitions for the generated RF scores.*

  - (**Adım 1:** *Oluşturulan RF skorları için segment tanımlamaları yapınız.*)

---

- **Step 2:** *Convert the scores into segments with the help of the matching below.*

  - (**Adım 2:** *Aşağıdaki eşleştirme yardımı ile skorları segmentlere çeviriniz.*)
 


---

```python
seg_map = {
    r'[1-2][1-2]' : 'hibernating',
    r'[1-2][3-4]' : 'at_risk',
    r'[1-2]5' : 'cant_loose_them',
    r'3[1-2]' : 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]' : 'loyal_customers',
    r'41': 'promising',
    r'[4-5][2-3]' : 'potential_loyalists',
    r'51' : 'new_customers',
    r'5[4-5]' : 'champions'
}
```

```python
rfm['segment'] = rfm['RF_SCORE'].replace(seg_map, regex=True)
```

## Preparation of the Entire Process with a Single Function (Tüm Sürecin Tek Fonksiyonla Hazırlanması)

```python
def rfm_preparation(dataframe, csv1=False, csv2=False, excel1=False, excel2=False):
    
    # Data Preparation (Veriyi Hazırlama)
    dataframe['omnichannel_order'] = dataframe['order_num_total_ever_online'] + dataframe['order_num_total_ever_offline']
    dataframe['omnichannel_value'] = dataframe['customer_value_total_ever_offline'] + dataframe['customer_value_total_ever_online']

    date_col = [col for col in dataframe.columns if 'date' in col]
    for col in date_col:
        dataframe[col] = pd.to_datetime(dataframe[col])

    # Calculation of RFM Metrics (RFM Metriklerinin Hesaplanması)
    rfm = pd.DataFrame()
    df['last_order_date'].max()
    today_date = dt.datetime(2021, 6, 1)
    rfm['customer_id'] = df['master_id']
    rfm['recency'] = (today_date - df['last_order_date']).astype('timedelta64[ns]').dt.days
    rfm['frequency'] = df['omnichannel_order']
    rfm['monetary'] = df['omnichannel_value']

    # Calculation of RF Score (RF Puanının Hesaplanması)
    rfm['recency_score'] = pd.qcut(rfm['recency'], 5, labels = [5, 4, 3, 2, 1])
    rfm['frequency_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 5, labels = [1, 2, 3, 4, 5])
    rfm['monetary_socre'] = pd.qcut(rfm['monetary'], 5, labels = [1, 2, 3, 4, 5])
    rfm['RF_SCORE'] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))

    # (Determination of Segments) Segmentlerin Belirlenmesi
    seg_map = {
        r'[1-2][1-2]' : 'hibernating',
        r'[1-2][3-4]' : 'at_risk',
        r'[1-2]5' : 'cant_loose_them',
        r'3[1-2]' : 'about_to_sleep',
        r'33': 'need_attention',
        r'[3-4][4-5]' : 'loyal_customers',
        r'41': 'promising',
        r'[4-5][2-3]' : 'potential_loyalists',
        r'51' : 'new_customers',
        r'5[4-5]' : 'champions'
    }
    
    rfm['segment'] = rfm['RF_SCORE'].replace(seg_map, regex=True)

    # First Task to Do (Yapılacak ilk görev)
    champ_loyal_id = rfm[(rfm['segment'] == 'champions') | 
                         (rfm['segment'] == 'loyal_customers')]['customer_id']

    women_champ_loyal_id = df[df['interested_in_categories_12'].str.contains('KADIN') & 
                              (df['master_id'].isin(champ_loyal_id))]['master_id']

    if csv1:
        women_champ_loyal_id.to_csv('womens_shoes_new_brand_target_customer_ids.csv', index=False)

    if excel1:
        women_champ_loyal_id.to_excel('womens_shoes_new_brand_target_customer_ids.xlsx', index=False)

    # Second Task to Do (Yapılacak İkinci görev)
    cant_loose_hiber_new_cust_id = rfm[((rfm['segment'] == 'cant_loose_them') | 
                                       (rfm['segment'] == 'hibernating') | 
                                       (rfm['segment'] == 'new_customers'))]['customer_id']

    man_child_cant_loose_hiber_new_cust_id = df[((df["interested_in_categories_12"].str.contains("ERKEK")) | 
                                                (df["interested_in_categories_12"].str.contains("COCUK"))) & 
                                                (df['master_id'].isin(cant_loose_hiber_new_cust_id))]["master_id"]

    if csv2:
        man_child_cant_loose_hiber_new_cust_id.to_csv('men_children_target_discount.csv', index=False)

    if excel2:
        man_child_cant_loose_hiber_new_cust_id.to_excel('men_children_target_discount.xlsx', index=False)

    return rfm
```
